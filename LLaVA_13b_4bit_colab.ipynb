```python
import pandas as pd
import requests
import os

df = pd.read_csv('/content/With_Value_West_Whately_01171005_processed.csv')

if not os.path.exists('downloaded_images'):
    os.makedirs('downloaded_images')

# 下载并保存图片
for index, row in df.iterrows():
    url = row['url']
    filename = row['filename']

    response = requests.get(url)

    if response.status_code == 200:
        with open(os.path.join('downloaded_images', filename), 'wb') as file:
            file.write(response.content)
```

```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('/content/Cleaned_Data_for_West_Whately_01171005.csv')

# 移除"value"列为空的行
df = df.dropna(subset=['value'])

# 调整"time_stamp"列到年月日的精度
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

# 保存处理后的CSV文件
df.to_csv('/content/With_Value_West_Whately_01171005_processed.csv', index=False)
```
